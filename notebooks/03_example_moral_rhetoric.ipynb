# Example 3: Investigating moral rhetoric in parliamentary debates

```python
import gepadeu
import utils
import glob
import spacy
import pandas as pd
from spacy import displacy
```

### Step 1: Read and process input
##### In step 1, we read in the speeches and convert them to GePaDeU objects.

##### Specify input directory and read input file names.

```python
indir = '../data/speeches'
jfiles = glob.glob(indir + '/*.json')
speeches = []
```

##### Read input files and convert them to GePaDeU objects.

```python
for jfile in jfiles: 
    with open(jfile, 'r') as inf: 
        speeches.append(gepadeu.Gepadeu.from_json(inf.read())) 
```

##### Extract textual content for each speech.

```python
content_dict = utils.extract_content(speeches)
```

##### Extract metadata for each speech.

```python
meta_dict = utils.extract_metadata(speeches)
```

##### Filter the set of speeches, according to some keyword list.

Parameters: 
- speeches to be filtered
- list of keyword strings
- strict_match (set to True if you want exact match => token == keyword; set to False for regex search)
- ignore_case (set to True to ignore upper-/lower case)

We use the keyword regex from Simonsen & Widmann (2025) to identify speeches on the topic of immigration: 

immigr*, migrat*, migrant*, migrier*, einwander*, zuwander*, zugewander*, eingewander*, asyl*, flüchtling*, 
geflücht*, ausländ*, gastarbeit*

Speeches are considered to be relevant if they contain at least 3 regex matches (either for the same or for different regexes).

```python
keyword_list = ['immigr', 'migrat', 'migrant', 'migrier', 'einwander', 'zuwander', 'zugewander', 'eingewander', 'asyl', 'flüchtling', 'geflücht', 'ausländ', 'gastarbeit']
strict_match = False
ignore_case  = True
filtered_speeches = utils.filter_speeches_by_keywords(speeches, keyword_list, strict_match, ignore_case)
print("No. speeches before filtering: " + str(len(speeches)) + "; no. of filtered speeches: " + str(len(filtered_speeches)))
```

### Step 2: Extract speechact and moral foundation annotations.

```python
speechact_dict = utils.extract_speechacts(filtered_speeches)
moral_dict = utils.extract_moral_frames(filtered_speeches)
```

##### Extract all speechacts for a specific speechact tag. Here we use the speechact tag "Accusation".

```python
speechact_tag = 'Accusation'
accusations = [(doc_id, item) for doc_id in list(speechact_dict.keys()) for item in speechact_dict[doc_id][speechact_tag]]
```

##### Display annotations for all instances with label speechact_tag.

```python
annots_per_party = gepadeu.Gepadeu.display_speechact_annotations(accusations, speechact_tag, meta_dict, "display")
```

##### Print distribution of speechact_tag across parties.

```python
for party in annots_per_party:
    print(party, len(annots_per_party[party]))
```

##### Combine speechacts with moral foundations and display results.

```python
annot_dict = gepadeu.Gepadeu.display_annots(speechact_dict, moral_dict, meta_dict, speechact_tag, 'display')
```

```python
# print distribution of moral foundations across parties for the specified speechact_tag 
cols = utils.get_mf_labels()
df = pd.DataFrame.from_dict(annot_dict, orient='index')
df['total'] = df.sum(axis=1)
for col in cols: 
    df[col] = round((df[col]/df.total)*100, 1)
df 
```
