# Example 2: Keyword search in GePaDeU 

```python
import gepadeu
import utils
import glob
import spacy
import pandas as pd
from spacy import displacy 
```

### Read data and do some preprocessing.

#### Specify input directory and read input file names.

```python
indir = '../data/speeches'
jfiles = glob.glob(indir + '/*.json')
speeches = []
```

#### Read input files and convert them to Gepadeu objects

```python
for jfile in jfiles: 
    with open(jfile, 'r') as inf: 
        speeches.append(gepadeu.Gepadeu.from_json(inf.read())) 
```

#### Extract content for each speech

```python
content_dict = utils.extract_content(speeches)
```

#### Extract metadata for each speech

```python
meta_dict = utils.extract_metadata(speeches)
```

#### Search for specific keywords and return their token ids as a list

```python
# Replace 'Freiheit' with your own keyword
keyword = 'Soldaten'
keyword_ids, num_instances = utils.search_for(speeches, keyword)
```

```python
# Print the number of keyword instances found in the data
print("Found ", num_instances, "instances of <", keyword, ">")
```

#### Get all speech act annotations (if any) for the keywords

```python
annotations = utils.get_annotations_by_id(keyword_ids, speeches, ['SPEECHACT'])
```

##### Display sentences that include the keyword and show speech act annotations. Also print speaker name, party affiliation and date the speech was given.

```python
annot_dict = gepadeu.Gepadeu.display_keyword_with_annots(annotations, meta_dict, keyword, ['SPEECHACT'], 'display')
```

#### Now add moral foundation annotations and display the results.

```python
keyword_ids, num_instances = utils.search_for(speeches, keyword)
layers = ['SPEECHACT', 'MORAL']
# Get all annotations (if any) for the annotation layers and the specified word
annotations = utils.get_annotations_by_id(keyword_ids, speeches, layers)
```

```python
annot_dict = gepadeu.Gepadeu.display_keyword_with_annots(annotations, meta_dict, keyword, layers, 'display')
```

##### For each party, print frequency for each annotated label for this keyword. The last argument specifies the annotation layer to be displayed.

```python
df = utils.keyword_per_party(keyword, annot_dict, meta_dict, 'MORAL') 
df
```
